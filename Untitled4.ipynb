```python
import numpy as np
import matplotlib.pyplot as plt

# Potential functions for each case
def potential_case1(phi):
    m = 1.0  # Mass parameter
    return 0.5 * m**2 * phi**2

def potential_case2(phi):
    lam = 1.0  # Coupling parameter
    return lam * phi**4

def potential_case3(phi):
    V0 = 1.0  # Potential amplitude
    lam = 1.0  # Exponential parameter
    return V0 * np.exp(-lam * phi)

# Equations of motion for each case
def equation_of_motion_case1(phi, phi_dot, H, m):
    return phi_dot, -3 * H * phi_dot - m**2 * phi

def equation_of_motion_case2(phi, phi_dot, H, lam):
    V_phi = potential_case2(phi)
    sqrt_term = np.sqrt(np.abs(1 - lam * phi_dot**2))  # Use abs to handle negative values
    return phi_dot, lam * V_phi / sqrt_term - 3 * H * phi_dot

def equation_of_motion_case3(phi, phi_dot, H, alpha):
    V_phi = potential_case3(phi)
    phi_dot_squared = phi_dot**2
    dV_dphi = -potential_case3(phi)  # Derivative of the potential
    sqrt_term = np.sqrt(np.abs(1 - alpha * phi_dot_squared))  # Use abs to handle negative values
    return phi_dot, alpha * dV_dphi / sqrt_term - V_phi * sqrt_term + 3 * H * (alpha - V_phi * sqrt_term)

# Fourth-order Runge-Kutta method
def rk4_step(func, phi, phi_dot, H, param):
    h = 0.0007 # Step size (adjust as needed)
    k1_phi_dot, k1_phi_dot_dot = func(phi, phi_dot, H, param)
    k2_phi_dot, k2_phi_dot_dot = func(phi + 0.5 * h * k1_phi_dot, phi_dot + 0.5 * h * k1_phi_dot_dot, H, param)
    k3_phi_dot, k3_phi_dot_dot = func(phi + 0.5 * h * k2_phi_dot, phi_dot + 0.5 * h * k2_phi_dot_dot, H, param)
    k4_phi_dot, k4_phi_dot_dot = func(phi + h * k3_phi_dot, phi_dot + h * k3_phi_dot_dot, H, param)
    phi += h * (k1_phi_dot + 2 * k2_phi_dot + 2 * k3_phi_dot + k4_phi_dot) / 6
    phi_dot += h * (k1_phi_dot_dot + 2 * k2_phi_dot_dot + 2 * k3_phi_dot_dot + k4_phi_dot_dot) / 6
    return phi, phi_dot

# Time span and initial conditions
t_span = (0, 10)  # Start and end time
t_eval = np.linspace(t_span[0], t_span[1], 1000)  # Time points for evaluation
phi_0 = 0.8  # Initial value of phi
phi_dot_0 = 0.0  # Initial value of phi_dot

# Case 1: f(alpha) = 1/2, alpha = 0
m_case1 = 1.0  # Mass parameter for Case 1
H_case1 = 0.1  # Hubble parameter for Case 1

phi_case1 = phi_0
phi_dot_case1 = phi_dot_0
V_phi_case1 = []

for t in t_eval:
    V_phi_case1.append(potential_case1(phi_case1))
    phi_case1, phi_dot_case1 = rk4_step(equation_of_motion_case1, phi_case1, phi_dot_case1, H_case1, m_case1)

# Case 2: f(alpha) = 0, alpha = -1
lam_case2 = 1.0  # Coupling parameter for Case 2
H_case2 = 0.1  # Hubble parameter for Case 2

phi_case2 = phi_0
phi_dot_case2 = phi_dot_0
V_phi_case2 = []

for t in t_eval:
    V_phi_case2.append(potential_case2(phi_case2))
    phi_case2, phi_dot_case2 = rk4_step(equation_of_motion_case2, phi_case2, phi_dot_case2, H_case2, lam_case2)

# Case 3: f(alpha) = 1, alpha = 1
alpha_case3 = 1.0  # Alpha parameter for Case 3
H_case3 = 0.1  # Hubble parameter for Case 3

phi_case3 = phi_0
phi_dot_case3 = phi_dot_0
V_phi_case3 = []

for t in t_eval:
    V_phi_case3.append(potential_case3(phi_case3))
    phi_case3, phi_dot_case3 = rk4_step(equation_of_motion_case3, phi_case3, phi_dot_case3, H_case3, alpha_case3)

# Plot V(phi) as a function of time for each case
plt.figure(figsize=(8, 6))
plt.plot(t_eval, V_phi_case1, label='Case 1')
plt.plot(t_eval, V_phi_case2, label='Case 2')
plt.plot(t_eval, V_phi_case3, label='Case 3')
plt.xlabel('Time')
plt.ylabel('V(phi)')
plt.legend()
plt.grid(True)
plt.title('Potential V(phi) as a Function of Time')
plt.show()
```
